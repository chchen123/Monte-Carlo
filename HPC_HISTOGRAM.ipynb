```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
full = h5py.File('/home/chen/run_0102.h5', 'r')
```

```python
dataset_name = '/monte carlo'
evt_full = full[dataset_name]
print(len(evt_full))
```

```python
monte_carlo = np.empty((0,0))
for i in range(int(len(evt_full))):
    monte_carlo = np.append(monte_carlo, np.array(evt_full[str(i)]))
    
print(monte_carlo)
```

```python
plt.hist(monte_carlo, label='Monte-Carlo', bins=10, color='blue',range=(0,120)) #blue
plt.legend(loc=2)
plt.title('Before cut')
plt.xlabel('chi^2')
plt.ylabel('count')
plt.show()
```

```python
full = h5py.File('/home/chen/run_0137.h5', 'r')
```

```python
dataset_name = '/monte carlo'
evt_full = full[dataset_name]
print(len(evt_full))
```

```python
monte_carlo = np.empty((0,0))
for i in range(len(evt_full)):
    try:
        monte_carlo = np.append(monte_carlo, np.array(evt_full[str(i)]))
    except Exception:
        print('Event ' + str(i) + ' failed MC fitting')
        continue

print(monte_carlo)
```
